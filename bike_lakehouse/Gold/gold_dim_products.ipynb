# Transformation Logic

```python
query = """
SELECT
    ROW_NUMBER() OVER (ORDER BY pn.start_date, pn.product_number) AS product_key, -- Surrogate key
    pn.product_id,
    pn.product_number,
    pn.product_name,
    pn.category_id,
    pc.category,
    pc.subcategory,
    pc.maintenance_flag,
    pn.product_line,
    pn.start_date
FROM de_project.silver.crm_products pn
LEFT JOIN de_project.silver.erp_product_category pc
    ON pn.category_id = pc.category_id
"""
df = spark.sql(query)
```

# Writing Gold Table

```python
df.write.mode("overwrite").format("delta").saveAsTable("de_project.gold.dim_products")
```
